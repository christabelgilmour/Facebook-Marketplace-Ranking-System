```python
import pandas as pd
import numpy as np
import pickle
df = pd.read_csv("Products.csv", index_col=0)
df = df.dropna(axis=0)
df["price"] = df["price"].str.strip("£")
df["price"] = df["price"].str.replace(",","")
df["price"] = df["price"].astype("float64")
print(np.sort(df["category"].unique()))
```

```python

df.loc[df["category"].str.contains("Appliance"), 'category'] = "Appliances"
df.loc[df["category"].str.contains("Home & Garden"), 'category'] = "Home & Garden"
df.loc[df["category"].str.contains("Baby & Kids"), 'category'] = "Baby & Kids"
df.loc[df["category"].str.contains("Video Games & Consoles"), 'category'] = "Video Games & Consoles"
df.loc[df["category"].str.contains("Clothes, Footwear & Accessories"), 'category'] = "Clothes, Footwear & Accessories"
df.loc[df["category"].str.contains("Computers & Software"), 'category'] = "Computers & Software"
df.loc[df["category"].str.contains("DIY Tools & Materials"), 'category'] = "DIY Tools & Materials"
df.loc[df["category"].str.contains("Health & Beauty"), 'category'] = "Health & Beauty"
df.loc[df["category"].str.contains("Music, Films, Books & Games"), 'category'] = "Music, Films, Books & Games"
df.loc[df["category"].str.contains("Sports, Leisure & Travel"), 'category'] = "Sports, Leisure & Travel"
df.loc[df["category"].str.contains("Office Furniture & Equipment"), 'category'] = "Office Furniture & Equipment"
df.loc[df["category"].str.contains("Other Goods"), 'category'] = "Other Goods"
df.loc[df["category"].str.contains("Phones, Mobile Phones & Telecoms"), 'category'] = "Phones, Mobile Phones & Telecoms"
print(df["category"])
```

```python
df["category"] = df["category"].astype("category")
print(np.sort(df["category"].unique()))
print(df.head())
```

```python
encoder = {'Appliances': 0, 'Baby & Kids': 1, 'Clothes, Footwear & Accessories': 2,
 'Computers & Software': 3, 'DIY Tools & Materials': 4, 'Health & Beauty': 5,
 'Home & Garden': 6, 'Music, Films, Books & Games': 7,
 'Office Furniture & Equipment': 8, 'Other Goods': 9,
 'Phones, Mobile Phones & Telecoms': 10, 'Sports, Leisure & Travel': 11,
 'Video Games & Consoles': 12}
```

```python
decoder = {value: key for key, value in encoder.items()}
print(decoder)
```

```python
df["labels"]= df["category"].replace(encoder)
df.head()
```
